```python
import pandas as pd
import numpy as np
import scipy.optimize
```

```python
a = pd.read_csv('Новая таблица - Лист1 (1).csv')
a = a.fillna(0)
a = a.drop('Unnamed: 0', 1)
a
```

```python
a_min = a.min().to_numpy()
a_max = a.max().to_numpy()
V_bounds = [(a_min[i], a_max[i]) for i in range(len(a_min))]
```

```python
a = a.to_numpy()
```

```python
b = np.zeros(len(a)) 
c = np.zeros(len(a_min))
c[1], c[23] = -1, -1 #целевая функция 
```

```python
fluxes = scipy.optimize.linprog(c=c, A_eq = a, b_eq = b, bounds = V_bounds)
fluxes
```

```python
res = pd.DataFrame({'res' : fluxes.x.tolist()})
res
```
